<h1>Pre-Processing</h1>

```python
def remove_spaces_between_word(dataset):
    for col in dataset.columns:
        for i in range(len(dataset[col])):
            if type(dataset[col][i]) == str:
                dataset[col][i] = dataset[col][i].strip()  # will remove additional space
                if dataset[col][i] == 'foul_smell_of urine':
                    dataset[col][i] = 'foul_smell_ofurine'

                else:
                    if ' _' in dataset[col][i]:
                        ele_list = dataset[col][i].split(' _')
                    elif '_ ' in dataset[col][i]:
                        ele_list = dataset[col][i].split('_ ')
                    else:
                        ele_list = dataset[col][i].split()
                    
                    dataset[col][i] = "_".join(ele_list)
    return dataset
```

```python
new_df = remove_spaces_between_word(data)
new_df.head()
```

<h3>Encoding Symptom</h3>

```python
def encoding(dataset):
    for ind in data_severity.index:
        dataset = dataset.replace(data_severity['Symptom'][ind], data_severity['weight'][ind])
    dataset = dataset.fillna(0) # put empty cell to 0
    return dataset           
```

```python
ds = encoding(data)
ds.head()
```

```python
def still_str(dataset):
    str_set = set()
    for col in dataset.columns[1:]:
        for i in range(len(dataset[col])):
            if type(dataset[col][i]) == str:
                str_set.add(dataset[col][i]);
    print(str_set)

still_str(ds)
```

<h1>Model</h1>

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix ,classification_report, precision_score
from sklearn.ensemble import RandomForestClassifier
```

```python
X = ds.drop('Disease',axis=1)
Y = ds['Disease']

X_train,X_test,y_train,y_test = train_test_split(X,Y,shuffle=True,train_size=0.7)
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
prediction = rf.predict(X_test)
print("\nAccuracy:",accuracy_score(y_test,prediction)*100)
print("\nPrecision:",precision_score(y_test,prediction,average='macro')*100)
print("\nF1-Score:",f1_score(y_test,prediction,average='macro')*100)
print("\nClassification Report:\n",classification_report(y_test, prediction))
```
